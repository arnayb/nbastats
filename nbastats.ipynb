```python
import pandas as pd
data = pd.read_csv("/kaggle/input/2023-2024-nba-player-stats/2023-2024 NBA Player Stats - Regular.csv",sep =";", encoding='latin-1')
```

```python
#Display player stats

player_name = input("Enter a player's name: ")
player_name_lower = player_name.lower()

player_stats = data[data['Player'].str.lower() == player_name_lower]

if not player_stats.empty:
    print("Here are the stats for", player_name + ":")
    # Display specific columns
    print(player_stats[['PTS', 'AST', 'TRB', 'STL', 'BLK', 'FG%', '3P%', 'FT%']])
else:
    print("Player not found.")
```
